```r
raw_dataset_mm <- read.csv("maternal_and_infant_mortality.csv")
```

```r
raw_dataset_health <- read.csv("insurance_coverage.csv")
```

```r
library(dplyr)
```

```r
maternal_mortality <- raw_dataset_mm %>% 
                        select (Variable, Country, Year, Value) %>% 
                    filter(Variable == "Maternal mortality")
```

```r
head(maternal_mortality)
```

```r
health_coverage <- raw_dataset_health %>% 
    select(Country, Variable, Year, Measure, Value) %>%
    filter(Year > 2004, Variable == "Total health care", Measure == "% of total population") 
head(health_coverage)
```

```r
joined_df_mm<- inner_join(
    maternal_mortality,
  health_coverage,
    by = c("Country", "Year")
    )
#here we are producing the simple Maternal Mortality data frame
```

```r

model1 <- lm(Value.x ~ Value.y, data = joined_df_mm)
summary(model1)
#here is our simple Maternal Mortality regression
```

```r
chi1 <- chisq.test(joined_df_mm$Value.x, joined_df_mm$Value.y)
print(chi1)
```

```r
neonatal_mortality <- raw_dataset_mm %>% 
                        select (Variable, Country, Year, Value) %>% 
                    filter(Variable == "Neonatal mortality, No minimum threshold of gestation period or birthweight")
```

```r
joined_df_nnm <- inner_join(
   neonatal_mortality,
  health_coverage,
    by = c("Country", "Year")
    )
summary(joined_df_nnm)
#here we are producing the simple Neonatal Mortality data frame
```

```r
model2 <- lm(Value.x ~ Value.y, data = joined_df_nnm)
summary(model2)
#here is our simple Neonatal Mortality regression
```

```r
chi2 <- chisq.test(joined_df_nnm$Value.x, joined_df_nnm$Value.y)
print(chi2)
```

```r
raw_fertility <- read.csv("fertility.csv")
```

```r
raw_life_expectancy <- read.csv("life_expectancy.csv")
```

```r
filtered_fertility <- raw_fertility %>% 
                        select (Variable, Country, Year, Value) %>% 
                    filter(Variable == "Fertility", Year >"2004")
head(filtered_fertility)
```

```r
filtered_life_expectancy <- raw_life_expectancy %>% 
                        select (Variable, Country, Year, Value) %>% 
                    filter(Variable == "Females at birth", Year>"2004")
head(filtered_life_expectancy)
```

```r
joined_df_mm_withjustfertility<- inner_join(
    filtered_fertility,
    joined_df_mm,
     by = c("Country", "Year")
    )
return(joined_df_mm_withjustfertility)
```

```r
joined_df_mm_withcontrols<- inner_join(
    joined_df_mm_withjustfertility,
    filtered_life_expectancy,
     by = c("Country", "Year")
    )
return(joined_df_mm_withcontrols)
#here is our complex Maternal Mortality dataframe
```

```r
model3 <- lm(Value.x.x ~ Value.y + Value.x + Value.y.y, data = joined_df_mm_withcontrols)
summary(model3)

#here is our complex Maternal Mortality regression
```

```r
joined_df_nnm_withjustfertility <- inner_join(
   joined_df_nnm,
  filtered_fertility,
    by = c("Country", "Year")
    )
head(joined_df_nnm_withjustfertility)
```

```r
joined_df_nnm_withcontrols <- inner_join(
   joined_df_nnm_withjustfertility,
  filtered_life_expectancy,
    by = c("Country", "Year")
    )
head(joined_df_nnm_withcontrols)
#here is our complex Neonatal Mortality dataframe
```

```r
model4 <- lm(Value.x ~ Value.y + Value.x.x + Value.y.y, data = joined_df_nnm_withcontrols)
summary(model4)

#here is our complex Neonatal Mortality regression
```
